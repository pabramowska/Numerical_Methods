```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#definijemy funkcje rysujące funkcję w przedziałąch i obliczające pochodne
def plot_function(a,b):
    x = [x for x in np.arange(a,b,0.001)]
    y = [-0.1*x**4+0.8*x**3-0.6*x**2-2*x+1.5 for x in x]
    plt.plot(x,y)
    plt.axhline(0, c='k', alpha=0.3)
    plt.show()
    
def first_derivative(a,b,h):
    x = [x for x in np.arange(a,b,0.001)]
    y = [((-0.1*(x+h)**4+0.8*(x+h)**3-0.6*(x+h)**2-2*(x+h)+1.5)-(-0.1*x**4+0.8*x**3-0.6*x**2-2*x+1.5))/h for x in x]
    plt.plot(x,y)
    plt.show()

def second_derivative(a,b,h):  
    x = [x for x in np.arange(a,b,0.001)]
    y = [((-0.4*(x+h)**3+2.4*(x+h)**2-1.2*(x+h)-2)-(-0.4*(x)**3+2.4*(x)**2-1.2*(x)-2))/h for x in x]
    plt.plot(x,y)
    plt.show()
```

```python
plot_function(-3,6)
```

```python
#rozpatrujemy przedział <-1.6;-1.4>
#rysujemy wykres funkcji w celu sprawdzenia przebiegu
plot_function(-1.6,-1.4)
```

```python
#wyliczenie pochodnej funkcji z definicji w przedziale <-1.6;-1,4>
#na wykresie widoczne jest, że pochodna maleje
first_derivative(-1.6,-1.4,0.00001)
```

```python
#wyliczanie drugiej pochodnej
#na wykresie widoczne jest, że druga pochodna ma ujemny znak
#dobieramy, więc punkt -1.6
second_derivative(-1.6,-1.4,0.00001)
```

```python
#sprawdzenie zgodności znaków
def is_positive(x0):
    px0=-0.1*x0**4+0.8*x0**3-0.6*x0**2-2*x0+1.5
    p_sec_derx0= -1.2*x0**2+4.8*x0-1.2
    return px0*p_sec_derx0>0
```

```python
is_positive(-1.6)
```

```python
def newton_method(x0,e):
    px0=-0.1*x0**4+0.8*x0**3-0.6*x0**2-2*x0+1.5
    px_derx0=-0.4*(x0)**3+2.4*(x0)**2-1.2*(x0)-2
    x=x0-(px0/px_derx0)
    
    while abs(x-x0)>e:
        x0=x
        px0=-0.1*x0**4+0.8*x0**3-0.6*x0**2-2*x0+1.5
        px_derx0=-0.4*(x0)**3+2.4*(x0)**2-1.2*(x0)-2
        x=x0-(px0/px_derx0) 
    
    return x
```

```python
p1=newton_method(-1.6,0.0001)
```

```python
#kolejnym rozpatrywanym przedziałem jest <0.5,1>
plot_function(0.5,1)
```

```python
#pierwsza pochodna jest rosnąca
first_derivative(0.5,1,0.0001)
```

```python
#druga pochodna również jest rosnąca
#funkcja i druga pochoda przyjmują o wartości o tych samych znakach dla x0 = 0.5
second_derivative(0.5,1,0.0001)
```

```python
#sprawdzenie czy wartość funkcji x0 * wartość drugiej pochodnej x0 > 0
is_positive(0.5)
```

```python
p2 = newton_method(0.5,0.0001)
```

```python
#kolejny przedział <2,2.5>
plot_function(2,2.5)
```

```python
first_derivative(2,2.5,0.0001)
```

```python
second_derivative(2,2.5,0.0001)
```

```python
is_positive(2.5)
```

```python
p3 = newton_method(2.5, 0.0001)
p=[p1,p2,p3]
yp = [0,0,0]
```

```python
#zmiana formatu wyników
p1 = "{:.3f}".format(p1)
p2 = "{:.3f}".format(p2)
p3 = "{:.3f}".format(p3)
```

```python
x = [x for x in np.arange(-3,6,0.001)]
y = [-0.1*x**4+0.8*x**3-0.6*x**2-2*x+1.5 for x in x]
plt.plot(x,y)
plt.axhline(0, c='k', alpha=0.3)
plt.scatter(p,yp,c='k')
plt.annotate(p1,(p[0],yp[0]),arrowprops=dict(ArrowStyle='->'),xytext=(p[0]-1.3,yp[0]+5))
plt.annotate(p2,(p[1],yp[1]),arrowprops=dict(ArrowStyle='->'),xytext=(p[1]-1.3,yp[1]+6))
plt.annotate(p3,(p[2],yp[2]),arrowprops=dict(ArrowStyle='->'),xytext=(p[2]-1.1,yp[2]+5.5))
plt.title('p(x) = -0.1*x^4 + 0.8*x^3 - 0.6*x^2 - 2*x + 1.5')
plt.show()
```
