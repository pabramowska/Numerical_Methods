```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#f(x)=1/(1+e^-0.25)

def f(x):
    return 1/(1+np.exp(-0.25*x))

#wyznaczenie funkcji oryginalnej
a = -2
b = 3
x = np.array([x for x in np.arange(a,b,0.001)])
y = f(x)
plt.plot(x,y)
```

### Węzły Legandre'a Stopnień wielomianu 1 

```python
t0 = 0.57735
t1 = -0.57735
A0 = A1 = 1
x0 = (b-a)/2*t0+(b+a)/2
x1 = (b-a)/2*t1+(b+a)/2

S1 = (b-a)/2*(f(x0)*A0+f(x1)*A1)
plt.plot(x,y)
plt.scatter([x0,x1],[f(x0),f(x1)], c='r')
```
