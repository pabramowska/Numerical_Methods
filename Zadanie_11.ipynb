```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#f(x)=1/(1+e^-0.25)
def f(x):
    return 1/(1+np.exp(-0.25*x))


#wyznaczenie funkcji oryginalnej
a = -2
b = 3
x = np.arange(a,b,0.001)
y = f(x)
plt.plot(x,y)
plt.legend(['f(x)'])
```

```python
#funkcja wyliczająca współczynniki xi
def count_x(a,b,t):
    return (b-a)/2*t+(b+a)/2
```

```python
#wyznaczanie funkcji S
def S(ti,Ai):
    x = np.array([count_x(a,b,t) for t in ti])
    plt.scatter(x, f(x))
    plt.plot(x, f(x), alpha = 0.3)
    print((b-a)/2 * sum(Ai*f(x)))
```

### Węzły Legandre'a Stopnień wielomianu 1 

```python
t1 = np.array([-0.57735, 0.57735])
A1 = np.array([1,1])
plt.plot(x,y)
S(t1,A1)
plt.legend(['f(x)', 'S(f)'])
```

```python
t4 = np.array([-0.906180, -0.538469, 0, 0.538469, 0.906180])
A4 = np.array([0.236927, 0.478629, 0.568889, 0.478629, 0.236927])
plt.plot(x,y)
S(t4,A4)
plt.legend(['f(x)', 'S(f)'])
```

### rysowanie przebiegów funkcji

```python
plt.plot(x,y)
plt.scatter(x,y,alpha=0)
S(t1,A1)
S(t2,A2)
S(t3,A3)
S(t4,A4)
plt.legend(['f(x)', 'S1(f)','S2(f)','S3(f)','S4(f)'])
```
