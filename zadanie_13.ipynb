```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#funkcja wyliczająca punkty metodą Eulera
def f(y):
    return -0.3*y


def euler_ulepszona(a,b,h,y0):
    y = [y0]
    i = np.arange(11)
    Xi = a+i*h
    for index in range(1,11):
        y.append(y[index-1] + h*f(y[index-1] + (h/2)*y[index-1]))
    return Xi,y
```

```python
#definiowanie zmiennych i wyliczanie punktów
h1 = 0.5
h2 = 0.25
a = 0
b = 1
y0 = 2
x1,y1 = euler_ulepszona(a,b,h1,y0)
x2,y2 = euler_ulepszona(a,b,h2,y0)
y1_text = ["{:.3f}".format(y1[index]) for index,_ in enumerate(y1)]
y2_text = ["{:.3f}".format(y2[index]) for index,_ in enumerate(y2)]
```

## wykres dla h=0.5

```python
plt.plot(x1,y1)
plt.scatter(x1,y1)
plt.xlabel('Xi')
plt.ylabel('Yi')
plt.title('Ulepszona Metoda Eulera dla h = 0.5')
for index in range(len(y1)):
    plt.annotate(y1_text[index],(x1[index],y1[index]),arrowprops=dict(ArrowStyle='->'),xytext=(x1[index]+0.5,y1[index]+0.02))
```

## wykres dla h = 0.25

```python
plt.plot(x2,y2)
plt.scatter(x2,y2)
plt.xlabel('Xi')
plt.ylabel('Yi')
plt.title('Ulepszona Metoda Eulera dla h = 0.25')
for index in range(len(y2)):
    plt.annotate(y2_text[index],(x2[index],y2[index]),arrowprops=dict(ArrowStyle='->'),xytext=(x2[index]+0.2,y2[index]))
```
