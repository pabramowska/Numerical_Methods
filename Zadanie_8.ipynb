```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [x for x in np.arange(-3,6,0.001)]
y = [-0.1*x**4+0.8*x**3-0.6*x**2-2*x+1.5 for x in x]
```

```python
plt.plot(x,y)
plt.grid()
plt.show()
```

przewidywany przedział pierwiastków <-1.8;-1.4>

```python
#definiowanie funkcji bisekcji
def bisection(a,b,e):
    
    while abs(a-b)>=e:
        fa = -0.1*a**4+0.8*a**3-0.6*a**2-2*a+1.5
        fb = -0.1*b**4+0.8*b**3-0.6*b**2-2*b+1.5
        p = (a+b)/2
        fp = -0.1*p**4+0.8*p**3-0.6*p**2-2*p+1.5
        if fa*fp<0:
            b=p
        else:
            a=p
    return a
```

```python
p1 = bisection(-1.8,-1.4,0.001)
```

```python
x_small = [x for x in np.arange(0,2.5,0.001)]
y_small = [-0.1*x**4+0.8*x**3-0.6*x**2-2*x+1.5 for x in x_small]
```

```python
plt.plot(x_small,y_small)
plt.grid()
plt.show()
```

kolejne rozpatrywane przedziały:
<0.5,1> i 
<2,2.5>

```python
p2 = bisection(0.5,1,0.001)
p3 = bisection(2,2.5,0.001)
p=[p1,p2,p3]
yp=[0,0,0]
```

```python
#zmiana formatu wyników
p1 = "{:.3f}".format(p1)
p2 = "{:.3f}".format(p2)
p3 = "{:.3f}".format(p3)
```

```python
plt.plot(x,y)
plt.axhline(0, c='k', alpha=0.3)
plt.scatter(p,yp,c='k')
plt.annotate(p1,(p[0],yp[0]),arrowprops=dict(ArrowStyle='->'),xytext=(p[0]-1.3,yp[0]+5))
plt.annotate(p2,(p[1],yp[1]),arrowprops=dict(ArrowStyle='->'),xytext=(p[1]-1.3,yp[1]+6))
plt.annotate(p3,(p[2],yp[2]),arrowprops=dict(ArrowStyle='->'),xytext=(p[2]-1.1,yp[2]+5.5))
plt.title('p(x) = -0.1*x^4 + 0.8*x^3 - 0.6*x^2 - 2*x + 1.5')
plt.show()
```
